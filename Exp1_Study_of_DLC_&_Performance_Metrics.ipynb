```python
#1. Load Data

import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

d = pd.read_csv('/content/drive/MyDrive/NNDL Lab/diabetes.csv')

d.describe()
d.head()
```

```python
# Splitting the dataset

x = d.iloc[:, 0:8]
y = d['Outcome']

print(x)
print(y)
```

```python
# 2. Define Keras Model - Model Building

model = Sequential()
model.add(Dense(12, input_shape=(8,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

```python
# 3. Compile Keras Model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. Fit the model
model.fit(x, y, epochs=70, batch_size=10) #fit Keras Model on Dataset
```

```python
# 5. Evaluate Keras Model

# Fit Keras Model on the dataset without progress bars
model.fit(x, y, epochs=70, batch_size=10, verbose=0)

# Evaluate the model
accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy[1]*100))
```

```python
 # 6. Make Predictions

predictions = (model.predict(x)>0.5).astype(int)
print(predictions)
```

```python
# Accuracy of Predictions

from sklearn.metrics import accuracy_score
accuracy_score(y, predictions)
```
